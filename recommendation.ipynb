```python
import pandas as pd
import warnings
import tkinter as tk

warnings.filterwarnings('ignore')
```

```python
app = tk.Tk()
app.title("Movie recommadation")
app.geometry('600x400')
app.config(bg='white')


u_data_col = ['user id' , 'item id' , 'rating' , 'timestamp']
u_data=pd.read_csv("u.data",sep='\t',names=u_data_col)

u_item_col = ['item id' , 'movie title' , 'release date', 'video release date',
             ' IMDb URL' , 'unknown' , 'Action' , 'Adventure' , 'Animation' ,
              "Children's" , 'Comedy' , 'Crime' , 'Documentary' , 'Drama' , 'Fantasy',
              'Film-Noir' ,'Horror' , 'Musical' ,'Mystery' , 'Romance' , 'Sci-Fi',
              'Thriller', 'War' , 'Western']
u_item=pd.read_csv('u.item',sep='|',encoding='latin-1',names=u_item_col)
u_item=u_item[['item id','movie title']]
data = pd.merge(u_data,u_item)
rating = pd.DataFrame(data.groupby('movie title')['rating'].mean())
rating['count'] = data['movie title'].value_counts()
pivot_df = data.pivot_table(index = 'user id', columns = 'movie title', values = 'rating')



f= tk.Frame(app,width=560,height=350,bd=5,bg='#FFFFE0',relief="ridge").place(x=20,y=25)
tk.Label(f,text='Movie you watched : ',font=("quiche sans", 12),bg='#FFFFE0',fg='orange').place(x=60,y=100)
tk.Label(f,text='Movie recommended : ',font=("quiche sans", 12),bg='#FFFFE0',fg='orange').place(x=60,y=260)
tk.Label(f,text='Movie you can try : ',font=("quiche sans", 12),bg='#FFFFE0',fg='orange').place(x=60,y=310)

title =tk.Label(f, text="Movie Recommendation System", font=("quiche sans", 15, "bold"), bg="#033054",
                      fg="white").place(x=50, y=35, width=500, height=35)

#create a variable to take user input, movie watched

movie_var = tk.Variable(app)
tk.Entry(f,textvariable=movie_var,).place(x=240,y=100,width=280,height=25)

#create a variable to receive recommended movie

recommend_var = tk.Variable(app)
recommend_var.set('Movie not entered , Please type....')
tk.Label(f,textvariable=recommend_var,bg='white',font=('quiche sans',10)).place(x=240,y=260,width=280,height=30)

#Create a variable t o receive also try movie

try_var = tk.Variable(app)
try_var.set('Movie not entered , Please type....')
tk.Label(f,textvariable=try_var,bg='white',font=('quiche sans',10)).place(x=240,y=310,width=280,height=30)


def find_recommend():
    movie = movie_var.get().lower().strip()
    if movie:
        try:
            movie = u_item['movie title'][u_item['movie title'].apply(lambda x : movie in x.lower())].values[0]
            movie_var.set(movie)
        except IndexError:
            recommend_var.set('Movie not found')
            try_var.set('Movie not found')
        else:
            corr_df = pd.DataFrame(pivot_df.corrwith(pivot_df[movie]),columns=['correlation'])
            corr_df.dropna(inplace=True)
            corr_df= corr_df.join(rating['count'])
            recommend_var.set(corr_df[(corr_df['count']>200) & (corr_df['correlation']>0.4)].sort_values(by='correlation',ascending=False).index[1])
            try_var.set(corr_df[corr_df['correlation']>0.4].sort_values(by='correlation',ascending=False).index[0])
    
            
def update():
    movie_var.set("")
    recommend_var.set("Movie not entered , Please type....")
    try_var.set("Movie not entered , Please type....")
    


tk.Button(f,text='Recommend',command=find_recommend,font=('quiche sans',12,'bold'),bg='#f44336',fg='white').place(x=160,y=186,width=130,height=30)
tk.Button(f,text='Update',command = update,font=('quiche sans',12,'bold'),bg='#f44336',fg='white').place(x=290,y=186,width=130,height=30)





app.mainloop()
```
