```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore') 
import tkinter as tk
from tkinter import ttk
```

```python
model = pd.read_pickle('Flight_Price_Predictor')
```

```python
import math

root=tk.Tk()
root.config(bg="white")
root.geometry('600x500')
root.title("Flight Price Predictor")

tk.Label(root,text='Flight Price Predictor',bg='#033054',fg='White',font=('goudy old style',15,'bold')).place(x=0,y=0,relwidth=1)
tk.Label(root,width=15,text="Name of Airline",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=40)
tk.Label(root,width=15,text="Source",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=70)
tk.Label(root,width=15,text="Destination",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=100)
tk.Label(root,width=15,text="Total Stops",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=130)
tk.Label(root,width=15,text="Additional Info :",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=160)
tk.Label(root,width=15,text="Date:",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=190)
tk.Label(root,width=15,text="Month:",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=220)
tk.Label(root,width=15,text="Depuarture Hour:",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=250)
tk.Label(root,width=15,text="Depaurture Min:",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=280)
tk.Label(root,width=15,text="Arrival Hour:",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=310)
tk.Label(root,width=15,text="Arrival Minute:",fg="black",bg="white",font=('goudy old style',15,'bold')).place(x=10,y=340)
tk.Label(root,width=15,text="Flight Ticket Price :",bg="light blue",fg='green',font=("Helvetica", 15,"bold")).place(x=20,y=450)


airline=tk.Variable(root)
source=tk.Variable(root)
Destination=tk.Variable(root)
stops=tk.Variable(root)
info=tk.Variable(root)
date=tk.Variable(root)
month=tk.Variable(root)
dep_hr=tk.Variable(root)
dep_min=tk.Variable(root)
arr_hr=tk.Variable(root)
arr_min=tk.Variable(root)

ttk.Combobox(root,width=25,textvariable=airline,font=('Arial',10),values=('IndiGo','Air India', 'Jet Airways','SpiceJet' ,'Multiple carriers', 'GoAir'
 'Vistara','Air Asia','Vistara Premium economy','Jet Airways Business',
 'Multiple carriers Premium economy','Trujet')).place(x=200,y=40)
ttk.Combobox(root,width=25,textvariable=source,font=('Arial',10),values=('Banglore','Kolkata','Delhi','Chennai','Mumbai')).place(x=200,y=70)
ttk.Combobox(root,width=25,textvariable=Destination,font=('Arial',10),values=('New Delhi','Banglore','Cochin','Kolkata','Delhi','Hyderabad')).place(x=200,y=100)
ttk.Combobox(root,width=25,textvariable=stops,font=('Arial',10),values=('0','1','2','3','4')).place(x=200,y=130)
ttk.Combobox(root,width=25,textvariable=info,font=('Arial',10),values=('1 Long layover','1 Short layover','2 Long layover','Business class','Change airports	','In-flight meal not included','No Info','No check-in baggage included','No info','Red-eye flight')).place(x=200,y=160)
tk.Entry(root,width=28,textvariable=date,font=('Arial',10),bg='light yellow').place(x=200,y=190)
tk.Entry(root,width=28,textvariable=month,font=('Arial',10),bg='light yellow').place(x=200,y=220)
tk.Entry(root,width=28,textvariable=dep_hr,font=('Arial',10),bg='light yellow').place(x=200,y=250)
tk.Entry(root,width=28,textvariable=dep_min,font=('Arial',10),bg='light yellow').place(x=200,y=280)
tk.Entry(root,width=28,textvariable=arr_hr,font=('Arial',10),bg='light yellow').place(x=200,y=310)
tk.Entry(root,width=28,textvariable=arr_min,font=('Arial',10),bg='light yellow').place(x=200,y=340)

predict_var = tk.Variable(root)
tk.Label(root,textvariable=predict_var,fg="crimson",bg='white',font=("Helvetica", 15,"bold")).place(x=300,y=450)

def predict():
    a=airline.get()
    b=source.get()
    c=Destination.get()
    d=stops.get()
    e=info.get()
    f=date.get()
    g=month.get()
    h=dep_hr.get()
    i=dep_min.get()
    j=arr_hr.get()
    k=arr_min.get()
    query = pd.DataFrame({'Airline':[a],'Source':[b],
                      'Destination':[c],'Total_Stops':[d],
                       'Additional_Info':[e],'Date':[f],
                       'Month':[g],'Dep_hour':[h],'Dep_minute':[i],'Arrival_hour':[j],'Arrival_minute':[k]})
    e=model.predict(query)[0]
    
tk.Button(root,text="Predict Price",command=predict,bg="midnightblue",fg="white",font=('Times new Roman',13,'bold'),cursor='hand2').place(x=200,y=380)
    

root.mainloop()
```
